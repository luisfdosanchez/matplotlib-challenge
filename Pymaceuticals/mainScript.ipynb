```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "rawdata/mouse_drug_data.csv"
clinical_trial_data_to_load = "rawdata/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouseDB=pd.read_csv(mouse_drug_data_to_load)
trialDB=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
dataDB=pd.DataFrame.merge(mouseDB,trialDB,on="Mouse ID")

# Display the data table for preview
dataDB.head()
```

### Lists for Iterations

```python
#Generating drug list (for iteration)
druglist=['Capomulin','Infubinol','Ketapril','Placebo']
colorlist=['b','g','r','k']
markerlist=['o','D','s','x']
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
tumor_mean=dataDB.groupby(["Drug", "Timepoint"]).mean()
tumor_mean=tumor_mean.drop(["Metastatic Sites"],axis=1)
tumor_mean=tumor_mean.reset_index()

# Preview DataFrame
tumor_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
tumor_sd=dataDB.groupby(["Drug", "Timepoint"]).sem()
tumor_sd=tumor_sd.drop(["Metastatic Sites", "Mouse ID"],axis=1)
tumor_sd=tumor_sd.reset_index()


# Preview DataFrame

tumor_sd.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#tumor_mean.head()
tumor_mean=tumor_mean.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
tumor_sd=tumor_sd.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
# Preview that Reformatting worked
tumor_sd.head()
tumor_mean.head()
```

```python
#Reset indices in DFs for easier manipulation
tumor_mean=tumor_mean.reset_index()
tumor_sd=tumor_sd.reset_index()
```

```python
# Generate the Plot (with Error Bars)

for x in range(len(druglist)):
    plt.errorbar(tumor_mean['Timepoint'], tumor_mean[druglist[x]], tumor_sd[druglist[x]], 
                 linestyle="dashed", linewidth=0.5, color=colorlist[x], marker=markerlist[x], markersize=4, 
                 label=druglist[x])
    
plt.legend(loc="upper left", fontsize='small')
plt.xlabel("Time (Days)")
plt.ylabel("Tumor volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.grid()
# Save the Figure
plt.savefig("./TumorResponse.png")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
# Convert to DataFrame
metastasis_mean=dataDB.groupby(["Drug", "Timepoint"]).mean()
metastasis_mean=metastasis_mean.drop(["Tumor Volume (mm3)"],axis=1)
metastasis_mean=metastasis_mean.reset_index()

# Preview DataFrame
metastasis_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
metastasis_sd=dataDB.groupby(["Drug", "Timepoint"]).sem()
metastasis_sd=metastasis_sd.drop(["Tumor Volume (mm3)", "Mouse ID"],axis=1)
metastasis_sd=metastasis_sd.reset_index()

# Preview DataFrame
metastasis_sd.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastasis_mean=metastasis_mean.pivot(index="Timepoint",columns="Drug",values="Metastatic Sites")
metastasis_sd=metastasis_sd.pivot(index="Timepoint",columns="Drug",values="Metastatic Sites")
# Preview that Reformatting worked
metastasis_mean.head()
metastasis_sd.head()
```

```python
#Reset indices in DFs for easier manipulation
metastasis_sd=metastasis_sd.reset_index()
metastasis_mean=metastasis_mean.reset_index()
```

```python
# Generate the Plot (with Error Bars)
for x in range(len(druglist)):
    plt.errorbar(metastasis_mean['Timepoint'], metastasis_mean[druglist[x]], metastasis_sd[druglist[x]], 
                 linestyle="dashed", linewidth=0.5, color=colorlist[x], marker=markerlist[x], markersize=4, 
                 label=druglist[x])
    
plt.legend(loc="upper left", fontsize='small')
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.title("Metastatic Spread During Treatment")
plt.grid()
# Save the Figure
plt.savefig("./MetastaticSpread.png")
plt.show()


```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
mouse_surv=dataDB.groupby(["Drug", "Timepoint"]).count()
mouse_surv=mouse_surv.drop(["Tumor Volume (mm3)", "Metastatic Sites"],axis=1)
mouse_surv=mouse_surv.rename(columns={"Mouse ID":"Mouse Count"})
mouse_surv=mouse_surv.reset_index()
mouse_survrate=mouse_surv
mouse_survrate["Mouse Count"]=mouse_survrate["Mouse Count"]*100/25
mouse_survrate=mouse_survrate.rename(columns={"Mouse Count":"Mouse Survival Rate"})

# Preview DataFrame
mouse_surv.head()
mouse_survrate.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_surv=mouse_surv.pivot(index="Timepoint",columns="Drug",values="Mouse Count")
mouse_survrate=mouse_survrate.pivot(index="Timepoint",columns="Drug",values="Mouse Survival Rate")
# Preview the Data Frame
mouse_survrate.head()
```

```python
#Reset indices in DFs for easier manipulation
mouse_surv=mouse_surv.reset_index()
mouse_survrate=mouse_survrate.reset_index()
```

```python
# Generate the Plot (Accounting for percentages)

for x in range(len(druglist)):
    plt.plot(mouse_survrate['Timepoint'], mouse_survrate[druglist[x]], 
                 linestyle="dashed", linewidth=0.5, color=colorlist[x], marker=markerlist[x], markersize=4, 
                 label=druglist[x])
    
plt.legend(loc="lower left", fontsize='small')
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")
plt.grid()
plt.savefig("./SurvivalRate.png")
plt.show()

```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
pctchg=[]
drugs=[]
for x in range(len(tumor_mean)+1):
    if x!=0:
        pctchg.append(100*((tumor_mean.iloc[9,x]/tumor_mean.iloc[0,x])-1))
        drugs.append(tumor_mean.iloc[:,x].name)

pctchg_DB=pd.DataFrame({"Drug":drugs,"Tumor % change":pctchg})
pctchg_DB.head(10)
# Display the data to confirm
```

```python
# Store all Relevant Percent Changes into a Tuple
pctchgSmall=[]
xticks=[.5,1.5,2.5,3.5]
for x in druglist:
    for y in range(len(pctchg_DB)):
        if x==pctchg_DB.iloc[y,0]:
            pctchgSmall.append(pctchg_DB.iloc[y,1])

pctchgtuple=(pctchgSmall)
pctchgtuple
           
    
# Splice the data between passing and failing drugs

# Orient widths. Add labels, tick marks, etc. 
for x in range(len(pctchgtuple)):
    if pctchgtuple[x]<0:
        plt.bar(druglist[x],pctchgtuple[x],width=1,color='green')
        plt.text(xticks[x]-.75,-10,str(round(pctchgtuple[x],1))+"%",size="large",color="white")
    else:
        plt.bar(druglist[x],pctchgtuple[x],width=1,color='red')
        plt.text(xticks[x]-.75,+7.5,str(round(pctchgtuple[x],1))+"%",size="large",color="white")
        
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.ylim(-30,70)
plt.xlim(-.75,3.75)
plt.grid()
plt.xticks(xticks,druglist)
# Save the Figure
plt.savefig("./TumorPctRate.png")
plt.show()
```
